## PySCF
Python Module for Quantum Chemistry.

Documentation: https://sunqm.github.io/pyscf/

Code: https://github.com/pyscf/pyscf

This driver can be used out-of-the-box, that is, PySCF does not need to be installed separately. It's features are available completely through Qiskit.

```python
from qiskit.chemistry.drivers import PySCFDriver, UnitsType

# Use PySCF, a classical computational chemistry software
# package, to compute the one-body and two-body integrals in
# molecular-orbital basis, necessary to form the Fermionic operator
driver = PySCFDriver(atom='H .0 .0 -0.25; H .0 .0 0.25',
                     unit=UnitsType.ANGSTROM,
                     basis='ccpvtz',
                     hf_method='rhf')
qmolecule = driver.run()
## one and two-body integrals
h1 = qmolecule.one_body_integrals
h2 = qmolecule.two_body_integrals
```

## PSI 4

Documentation: http://www.psicode.org/psi4manual/1.3.2/index.html

**To use this driver, the PSi4 software needs to be installed.** It's not installed where this notebook is running, but example usage is shown below.


Qiskit usage example here: 
    https://github.com/Qiskit/qiskit-community-tutorials/blob/f589e0aa4b31de30aed01f8219abdd3e56771653/chemistry/h2_basis_sets.ipynb

```python
from qiskit.chemistry.drivers import PSI4Driver
# PSI4 config here is a multi-line string that we update using format()
# To do so all other curly brackets that are required in the PSI4 config must be doubled
psi4_cfg = """
molecule h2 {{
   0 1
   H 0.0 0.0 -{0}
   H 0.0 0.0 {0}
}}

set {{
  basis {1}
  scf_type pk
}}
"""
basis_sets = ['sto-3g', '3-21g', '6-31g']
driver = PSI4Driver(psi4_cfg.format(0.25, basis_sets[0]))
```
